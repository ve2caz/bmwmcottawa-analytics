```python
import pandas as pd  # type: ignore


def create_reference(df):
    # Generated by Copilot using DataWrangler
    column_descriptions = {
        "Event Date": "Date of the event",
        "Event Start Time": "Start time of the event",
        "Event Finish Time": "Finish time of the event",
        "Event Category": "Category of the event",
        "Event Title": "Title of the event",
        "Registrant Last Name": "Last name of the registrant",
        "Registrant First Name": "First name of the registrant",
        "Member Number": "Membership number of the registrant",
        "Registrant Type": "Type of the registrant",
        "Registration Status": "Status of the registration",
        "Attended": "Attendance status",
    }
    # Convert to DataFrame
    df = pd.DataFrame(list(column_descriptions.items()), columns=["Column Name", "Description"])
    return df


events_and_registrants_df = pd.read_csv(r"../data/raw/events_and_registrants.csv")
events_and_registrants_schema_df = create_reference(events_and_registrants_df.copy())
```

```python
title = "# Events and Registrants Schema\n\n"
markdown_output = title + events_and_registrants_schema_df.to_markdown(index=False)
with open("../references/03-EventsAndRegistrants-Schema.md", "w") as file:
    file.write(markdown_output)
```
